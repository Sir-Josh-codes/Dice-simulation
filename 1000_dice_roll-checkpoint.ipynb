```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#simulating 1000 dice rolls
rolls= np.random.randint(1,7, size= 1000)
# conversion to pandas series for easier analysis
rolls_series= pd.Series(rolls)
```

```python
# counting occurrence of each number
counts = rolls_series.value_counts().sort_index()
# calculating probabilities
probabilities = counts / 1000
print ('Counts:\n', counts)
print('\nProbabilities:\n', probabilities)
```

```python
# visualization of results
plt.bar(counts.index, probabilities)
plt.xlabel('Dice Face')
plt.ylabel('Probability')
plt.title('Probability distribution of 1000 dice rolls')
plt.show()
```
